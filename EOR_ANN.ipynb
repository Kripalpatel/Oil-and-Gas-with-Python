```
# -*- coding: utf-8 -*-
"""
Created on Wed Aug 26 18:05:20 2020

@author: Kipu
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
#from ann_visualizer.visualize import ann_viz

def generate_api(mode, min, max, n):
  np.random.seed(10)
  return(np.random.triangular(min, mode, max, n))

def generate_visco(mode, min, max, n):
  np.random.seed(10)
  return(np.random.triangular(min, mode, max, n))

def generate_sat(mode, min, max, n):
  np.random.seed(10)
  return(np.random.triangular(min, mode, max, n))

def generate_perm(mode, min, max, n):
  np.random.seed(10)
  return(np.random.triangular(min, mode, max, n))

def generate_dept(mode, min, max, n):
  np.random.seed(10)
  return(np.random.triangular(min, mode, max, n))

def generate_temp(mode, min, max, n):
  np.random.seed(10)
  return(np.random.triangular(min, mode, max, n))

# Model for the N2 
  
api_n2 = generate_api(48, 35, 55, 35)
visco_n2 = generate_visco(0.2, 0.01, 0.4, 35)
sat_n2 = generate_sat(0.75, 0.4, 0.8, 35)
perm_n2 = generate_perm(2500, 1, 5000, 35)
dept_n2 = generate_dept(8000, 6000, 10000, 35)
temp_n2 = generate_temp(210, 95, 300, 35)
label_n2 = np.full(35, 'Nitrogen')

a = temp_n2

#print(np.mean(a))
plt.hist(a, bins=25)
#plt.show()

# for Light HC Flooding 

api_hc = generate_api(41, 24, 50, 35)
visco_hc = generate_visco(0.5, 0.01, 3, 35)
sat_hc = generate_sat(0.8, 0.3, 0.9, 35)
perm_hc= generate_perm(2500, 1, 5000, 35)
dept_hc = generate_dept(6000, 4000, 10000, 35)
temp_hc = generate_temp(210, 95, 300, 35)
label_hc = np.full(35, 'Light HC')

a2 = temp_hc

#print(np.mean(a))
plt.hist(a2, bins=25)
#plt.show()

# for Co2 Injection 

api_Co2 = generate_api(36, 23, 47, 35)
visco_Co2 = generate_visco(1.5, 0.01, 10, 35)
sat_Co2 = generate_sat(0.55, 0.2, 0.8, 35)
perm_Co2= generate_perm(2500, 1, 5000, 35)
dept_Co2 = generate_dept(7500, 2500, 10000, 35)
temp_Co2 = generate_temp(210, 95, 300, 35)
label_Co2 = np.full(35, 'Co2 Injection')

a3 = temp_Co2

#print(np.mean(a))
plt.hist(a3, bins=25)
#plt.show()

# for Immicible gas 

api_immgas = generate_api(22, 13, 37, 35)
visco_immgas = generate_visco(100, 1, 600, 35)
sat_immgas = generate_sat(0.7, 0.35, 0.9, 35)
perm_immgas = generate_perm(2500, 1, 5000, 35)
dept_immgas = generate_dept(5500, 1800, 9000, 35)
temp_immgas = generate_temp(210, 95, 300, 35)
label_immgas = np.full(35, 'Immiscible gas')

a4 = temp_immgas

#print(np.mean(a))
plt.hist(a4, bins=25)
#plt.show()

# alkaline flooding 

api_alka = generate_api(35, 20, 40, 35)
visco_alka = generate_visco(13, 1, 35, 35)
sat_alka = generate_sat(0.53, 0.35, 0.8, 35)
perm_alka = generate_perm(450, 10, 800, 35)
dept_alka = generate_dept(3250, 1000, 7000, 35)
temp_alka = generate_temp(80, 70, 150, 35)
label_alka = np.full(35, 'ASP')

a5 = temp_alka

#print(np.mean(a5))
plt.hist(a5, bins=25)
#plt.show()

# polymer flooding 

api_poly = generate_api(35, 15, 55, 35)
visco_poly = generate_visco(85, 1, 150, 35)
sat_poly = generate_sat(0.8, 0.5, 0.9, 35)
perm_poly = generate_perm(800, 20, 1700, 35)
dept_poly = generate_dept(4750, 500, 9000, 35)
temp_poly = generate_temp(140, 75, 170, 35)
label_poly = np.full(35, 'polymer flooding')

a6 = temp_poly

#print(np.mean(a6))
plt.hist(a5, bins=25)
#plt.show()

# steam flooding 

api_steam = generate_api(13.5, 8, 25, 35)
visco_steam = generate_visco(4700, 1, 200000, 35)
sat_steam = generate_sat(0.66, 0.4, 0.9, 35)
perm_steam = generate_perm(2540, 200, 5000, 35)
dept_steam = generate_dept(1500, 300, 4500, 35) 
temp_steam = generate_temp(210, 95, 300, 35) 
label_steam = np.full(35, 'Steam')

a7 = dept_steam

#print(np.mean(a))
plt.hist(a7, bins=25)
#plt.show()

# append the all data to the common veriable 

api = np.concatenate([api_n2, api_hc, api_Co2, api_immgas, api_alka, api_poly, api_steam], axis=0)
visco = np.concatenate([visco_n2, visco_hc, visco_Co2, visco_immgas, visco_alka, visco_poly, visco_steam], axis=0)
sat = np.concatenate([sat_n2, sat_hc, sat_Co2, sat_immgas, sat_alka, sat_poly, sat_steam], axis=0)
perm = np.concatenate([perm_n2, perm_hc, perm_Co2, perm_immgas, perm_alka, perm_poly, perm_steam], axis=0)
dept = np.concatenate([dept_n2, dept_hc, dept_Co2, dept_immgas, dept_alka, dept_poly, dept_steam], axis=0)
temp = np.concatenate([temp_n2, temp_hc, temp_Co2, temp_immgas, temp_alka, temp_poly, temp_steam], axis=0)
label = np.concatenate([label_n2, label_hc, label_Co2, label_immgas, label_alka, label_poly, label_steam], axis=0)

train_df = pd.DataFrame({'API':api, 'Viscosity':visco, 'Oil Saturation':sat, 'Depth':dept, 'Temperature':temp,'Label':label})
#print(train_df)

g = sns.pairplot(train_df, hue = 'Label');
g.fig.suptitle("Pairplot of 6 Reservoir Fluid and Rock Properties of 245 Simulated Observations (TRAIN)", size=20, y=1.05)
plt.show()

df_for_corr = train_df.iloc[:,0:-1]
corr = df_for_corr.corr()

mask = np.zeros(corr.shape, dtype=bool)
mask[np.triu_indices(len(mask))] = True

fig = plt.gcf()
fig.set_size_inches(12, 8)

g = sns.heatmap(corr, annot=True, vmin=-0.8, vmax=0.8, cmap='seismic', mask=mask)
plt.show()

# prediction data 

np.random.seed(10)

api_pred = np.random.uniform(min(api), max(api), size=(50,))
visco_pred = np.concatenate([(np.random.triangular(min(visco), 1, 3, 20)), (np.random.triangular(3, 10, 100, 20)), (np.random.uniform(100, 200000, size=(10,)))], axis=0)
sat_pred = np.random.uniform(min(sat), max(sat), size=(50,))
perm_pred = np.random.uniform(min(perm), max(perm), size=(50,))
dept_pred = np.random.uniform(min(dept), max(dept), size=(50,))
temp_pred = np.random.uniform(min(temp), max(temp), size=(50,))

pred_df = pd.DataFrame({'Density': api_pred, 'Viscosity': visco_pred, 'Oil Saturation': sat_pred, 
                         'Permeability': perm_pred, 'Depth': dept_pred, 'Temperature': temp_pred})
#print(pred_df)

# neural network 

X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1:].astype(str).values

X_pred = pred_df

from sklearn.preprocessing import OneHotEncoder

y_train = np.array(list(y_train)).reshape(-1,1)

oh = OneHotEncoder()
oh.fit(y_train)
y_train = oh.transform(y_train).todense()

#print(y_train.shape)
#print(X_pred.shape)

import keras
from keras.layers import Dense
from keras.models import Sequential

n_cols = X_train.shape[1]

model = Sequential()

# Add the first layer
model.add(Dense(132, activation='relu', input_shape=(n_cols,)))
model.add(Dense(132, activation='relu'))
model.add(Dense(132, activation='relu'))

# Add the output layer, 2 nodes because it's binary classification
model.add(Dense(7, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=600)

```
